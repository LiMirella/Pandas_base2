# Pandas I/O

```python
import pandas as pd
```

Importação de arquivo CVS

```python
url = 'https://raw.githubusercontent.com/alura-cursos/Pandas/refs/heads/main/clientes_banco.csv'
dados = pd.read_csv(url)
dados.head()
```

Salvando arquivo CVS

```python
dados.to_csv('dados.csv', index = False)
```

Importação de arquivos em xlsx, xls, xlsm, xlsb (excel)

```python
dados2 = pd.read_excel('/content/emissoes_CO2.xlsx') #importação com imput do arquivo no colab
dados2.head()
```

```python
url3 = 'https://github.com/alura-cursos/Pandas/blob/main/emissoes_CO2.xlsx?raw=True' #importação direto do github
dados3 = pd.read_excel(url3)
dados3.head()
```

Escolhendo qual pagina iremos visualizar

```python
pd.ExcelFile(url3).sheet_names #verificando o nome das paginas
```

```python
dados4 = pd.read_excel(url3, sheet_name='emissoes_percapita') #selecionando a pagina e salvando dentro de uma variavel que será exibida
dados4.head()
```

```python
dados4 = pd.read_excel(url3, sheet_name='emissoes_percapita', usecols='A:C') #o parametro usecols serve para selecionarmos as colunas que serão exibidas
dados4.head()
```

Salvando arquivo excel

```python
dados4.to_excel('emissoes_percapita.xlsx')
```
